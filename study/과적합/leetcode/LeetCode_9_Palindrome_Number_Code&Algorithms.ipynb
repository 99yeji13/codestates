# LeetCode 9. Palindrome Number

Given an integer x, return true if x is palindrome integer.

An integer is a palindrome when it reads the same backward as forward.

For example, 121 is a palindrome while 123 is not.

( Palindrome : 글자를 순서대로 읽어도, 역순으로 읽어도 같은 단어가 되는 것 )
- ex) "기러기", "토마토", "스위스", "인도인", "별똥별", "우영우", "역삼역"

=> 주어진 정수형 숫자가 Palindrome 숫자인지 아닌지 판단하여 그 결과를 true/ false로 반환하라

## Solutions

- 문자열로 바꿔서 비교
- 리스트

(* 대칭수인 Palindrome은 음수일 수 없다.)

### 문자열로 바꿔서 비교

```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        return False if x < 0 else str(x)[::-1] == str(x)
```

### 리스트

```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        if x < 0:
            return False
        nums = []
        while x > 0:
            x, y = x//10, x%10
            nums.append(y)
        return nums == nums[::-1]
```
